```python
df.dropna(subset=['Latitude', 'Longitude'], inplace=True)
```

```python
# Assualt, Robbery, Vandalism, Motor Vehicle theft, 
subset_df = df[(df['Incident Category'] == 'Assualt') | (df['Incident Category'] == 'Robbery') | (df['Incident Category'] == 'Vandalism') | (df['Incident Category'] == 'Motor Vehicle theft') | (df['Incident Description'] == 'GRAND THEFT FROM LOCKED AUTO')][['Incident Datetime','Incident Category','Incident Description','Latitude', 'Longitude']]
```

```python
subset_df.drop(['Distance', 'Distance2'], inplace=True, axis=1)
```

```python
subset_df['Incident Datetime'] = pd.to_datetime(subset_df['Incident Datetime'], format='%Y/%m/%d %I:%M:%S %p')
```

```python
from geopy.distance import geodesic
target_lat = 37.787087
target_lon = -122.421560

# Define the search radius (in kilometers)
search_radius = 5  # You can adjust this value as per your requirement

# Create a function to calculate the distance between two coordinates
def get_distance(lat1, lon1, lat2, lon2):
    return geodesic((lat1, lon1), (lat2, lon2)).km
```

```python
subset_df['Distance'] = subset_df.apply(lambda row: get_distance(row['Latitude'], row['Longitude'], target_lat, target_lon), axis=1)
```
